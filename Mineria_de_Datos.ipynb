```python
#Evalua el desempeño de la expresion respecto al tiempo, solo evalua una linea
%timeit L = [n ** 2 for n in range(1000)]
```

```python
#El doble porcentaje es para evaluar el desempeño de varias lineas de tiempo
%%timeit
L = []
for n in range(1000):
  L.append(n ** 2)
```

Understanding Data Types in IPython

```python
#Importa numpy y renombralo como np
import numpy as np
```

```python
#un arregl lineal de enteros, ND array, number dimention
np.array([1, 4, 2, 5, 3])
```

```python
#al hacer una lista con np, esta castea los diferentes numeros en un tipo, en este caso, flotantes
np.array([3.14, 4, 2, 3])
```

```python
#podemos especificar que tipo queremos castear
np.array([3.14, 4, 2, 3], dtype='float32')
```

```python
np.array([3.14, 4, 2, 3], dtype='int')
```

```python
#En rango no se toma en cuenta el indice superior, solo el inferior, lista de comprension anidada, el resultado es una matriz
np.array([range(i, i + 3) for i in [2, 4, 6]])
```
